```python
import urllib.request as r
from bs4 import BeautifulSoup
import re
import time
import pandas as pd 
import numpy as np 


naver_movie = 'https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=181381&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page=' 
script = []
id_list = []
star_list = []
good_list = []
bad_list = []
date_list = []


for no in range(1,1248):
	
	#if no == 218:
	#	pass	

	#elif no == 1090:
	#	pass

	if no == 1247:
		movie_url = naver_movie+str(no)
		f = r.urlopen(movie_url) # 만든 웹페이지 주소 불러오기 
		html = f.read()                    # 불러온 웹페이지 html코드 전부 읽기
		bs = BeautifulSoup(html, 'lxml')      #BeautifulSoup에 해당 불러온 웹페이지 html 코드 전부 넣고 bs 객체 생성 
		test = bs.find_all('li')
		
		for i in range(6,8):
			naid = test[i].select('em > a')[0].text[1:-1] ## 아이디
			socre = test[i].select('div.star_score')[0].select('em')[0].text ## 평점
			good = test[i].select('strong')[0].text ## 공감
			bad = test[i].select('strong')[1].text ## 비공감
			date = test[i].select('dt > em')[1].text ## 날짜
			id_list.append(naid)
			star_list.append(socre)
			good_list.append(good)
			bad_list.append(bad)
			date_list.append(date)
			try:
				a = test[i].select('p')[0].select('a')[0]['data-src']
				script.append(a)
			except:
				b = test[i].select('p')[0].text
				b = re.sub("\t|\n|\r","",b)
				script.append(b)
	
	elif no < 1247:
		movie_url = naver_movie+str(no)
		f = r.urlopen(movie_url) # 만든 웹페이지 주소 불러오기 
		html = f.read()                    # 불러온 웹페이지 html코드 전부 읽기
		bs = BeautifulSoup(html, 'lxml')      #BeautifulSoup에 해당 불러온 웹페이지 html 코드 전부 넣고 bs 객체 생성 
		test = bs.find_all('li')
		
		for i in range(6,16):
			naid = test[i].select('em > a')[0].text[1:-1] ## 아이디
			socre = test[i].select('div.star_score')[0].select('em')[0].text ## 평점
			good = test[i].select('strong')[0].text ## 공감
			bad = test[i].select('strong')[1].text ## 비공감
			date = test[i].select('dt > em')[1].text ## 날짜
			id_list.append(naid) ## 아이디 리스트에 추가
			star_list.append(socre) ## 점수 리스트에 추가
			good_list.append(good) ## 공감 리스트에 추가
			bad_list.append(bad) ## 비공감 리스트에 추가
			date_list.append(date) ## 날짜 리스트에 추가
			try: ## 접혀있는 리뷰 펴서 보기
				a = test[i].select('p')[0].select('a')[0]['data-src']
				script.append(a)
			except: ## 접혀있지 않을 경우 보기
				b = test[i].select('p')[0].text
				b = re.sub("\t|\n|\r","",b)
				script.append(b)


	print(no,'페이지를 파싱 중')
```
